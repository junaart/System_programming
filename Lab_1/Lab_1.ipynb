# Лабораторная работа 1. Рабочее окружение. Командная строка Linux.

## Знакомство с рабочим окружением

1. Создайте профиль на github
2. Внутри профиля создайте репозиторию System_programming public
3. Создайте в репозитории System_programming кодовое пространство codespase
4. В кодовом пространстве создайте рабочее окружение, установив gcc,  gdb, fasm, mc.
5. Создайте файл на языке ассемблера типа "Hello, world", только вместо этого выведите свои фамилию, имя, отчество (каждую с новой строки).
6. Сохраните файл с именем своей фамилии и расширением asm, создайте исполняемый файл для 32-разрядной ОС, убедитесь в его работе.
7. Создайте аналогичную программу на языке C, назвав его "С_version.c", скомпилируйте его с помощью gcc, проверьте работу.
8. Дизассемблируйте первый и второй исполняемые файлы с помощью gdb, записав результат дизассемблирования в текстовый файл "Disassemble".
9. Создайте папку sandbox и переместите туда все созданные файлы.

## Работа с файлами

**Все команды, вводимые при выполнении пунктов заданий, записывайте в файл "Work_files"**

1. В рабочем каталоге создайте создайте папку «Work», внутри которой еще две папки «Лабораторная работа №1!» «ФИО».
2. Установите на папку «Контрольная работа №1!» права для владельца и группы владельца: чтение, запись, выполнение, для остальных права запись; на папку «ФИО» для владельца права только на запись.
3. Создайте в папке «Work» три файла «Text @1», «Text $2», «Text  #3».
4. Без использования текстового редактора запишите в файл «Text @1» следующий текст: 

«Птица говорун отличается умом и сообразительностью!»

с новой строки: 

«Отличается умом, отличается сообразительностью...»

5. Без использования текстового редактора скопируете содержимое файла «Text @1» в файл «Text $2».

6. Переместите файл «Text $2» в папку «Лабораторная работа №1!» и скопировать файл «Text @1» в папку «ФИО».

7. Сожмите файлы «Text @1», «Text  #3» с помощью утилиты gzip и добавьте их в архив «MyAchiv» с помощью утилиты tar.

8. Добавьте в файл «Text  #3» первую строку из файла  «Text @1» без использования текстового редактора, а далее в конец «Text  #3» добавьте строку «Будь осторожен! Преступник вооружен!»

9. В папке «ФИО» создайте файл «Result_one», в который запишите в обратном порядке содержимое файла «Text @1»

10. Переименуйте файл «Result_one» в «Result_two»

11. В папке «Лабораторная работа №1!» создайте файл «Result_3», в который запишите содержимое файлов и папок папки «ФИО», начинающихся в своем названии с буквы «T».

12. Добавьте, без использования текстового редактора, в конец файла «Result_3» полную информацию о ядре системы и текущую дату.
