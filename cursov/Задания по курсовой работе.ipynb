# Требования к распределению вариантов курсовой работы

1. Можно работать в команде по два человека. Финальная версия программы должна отличаться различными вариантами расширения функциональности программы.
2. Можно предложить свое задание на согласование преподавателю с подробным описанием содержания по аналогии с банком заданий ниже.

# Требования к выполнению задания курсовой работы

1. Программы разрабатываются на языке ассемблера Fasm на платформе GNU/Linux. Можно использовать библиотеки на языке С (ncurses и т.д.)
2. Рабочее окружение может быть локальным или облачным (github/codespace).

# Требования по оформлению курсовой работы

Требования к содержанию и форме разрабатываемых документов изложены в указаниях кафедральной разработки, которые предоставляются студентам и размещены на сайте кафедры. Разрабатываемая в рамках курсовой работы документация должна включать:

- Техническое задание (ТЗ) на разработку программы резидента (техническое задание должно содержать, помимо общих, и конкретные пункты ТЗ для варианта задания).
- Описание применения (ОП) ПО.
- Техническое описание (ТО) программы ПО (описание всех процедур, их входных и выходных параметров, программ и структур данных, включая описание модульной структуры и блок-схем программ, которые выносятся на листы).
- Текст программы в виде листинга (ЛП).
- Руководство системного программиста (РСП) (в том числе обязательно: состав ПО, системные требования к ОС, используемые прерывания, требования к развертыванию и удалению программного обеспечения).
- Руководство пользователя (РП): все от "А" до "Я" по работе пользователя, включая инструкцию командной строки. Инструкция должна быть понятна пользователю, и ориентирована именно на такого пользователя, которому функционально предназначен программный продукт. Должны быть приведены примеры запуска и использования программы.
- Программа и методика испытаний (ПМИ) (основное - в виде таблицы для проверки всех пунктов ТЗ со ссылкой на них, обязательно условия проведения испытаний и результат выполнения пункта). Обязательно должны быть отражены название испытуемого изделия и программы, условия проведения испытаний, действия для проверки и результаты этих действий (вплоть до нажатия отдельных клавиш). ПМИ разрабатывается для отдельной проверки выполнения каждого пункта ТЗ и работоспособности.
- CD, DVD, или флеш-носитель с исходными текстами, документацией и резидентной программой готовой к выполнению, откомпилированной в fasm format ELF64 и слинкованной ld.

Документация оформляется в соответствии с ГОСТ. Все документы должны иметь отдельный титульный лист (форма титульного листа дана в документе – требования к документации), оформленный по требованиям кафедры.


# Банк заданий по курсовой работе   

**Вариант 1.** Игра в кости: играющий называет любое число в диапазоне от 2 до 12 и ставку, которую он делает в этот ход. Программа с помощью датчика случайных чисел дважды выбирает числа от 1 до 6 ("бросает кубик", на гранях которого цифры от 1 до 6). Если сумма выпавших цифр меньше 7 и играющий задумал число меньше 7, он выигрывает сделанную ставку. Если сумма выпавших цифр больше 7 и играющий задумал число больше 7, он также выигрывает сделанную ставку. Если играющий угадал сумму цифр, он получает в четыре раза больше очков, чем сделанная ставка. Ставка проиграна, если не имеет место ни одна из описанных ситуаций. В начальный момент у играющего 100 очков. Предложите и реализуйте расширения функциональности этой игры (например, 1. вместо человека играет другая программа, 2. казино копит проигранные ставки и т.п.).

**Вариант 2.** Игра в морской бой: на поле 10x10 клеток программа размещает флот кораблей: 1 корабль — ряд из 4 клеток («четырёхпалубный»; линкор); 2 корабля — ряд из 3 клеток («трёхпалубные»; крейсера); 3 корабля — ряд из 2 клеток («двухпалубные»; эсминцы); 4 корабля — 1 клетка («однопалубные»; торпедные катера). При размещении корабли не могут касаться друг друга сторонами и углами. Предложите и реализуйте  расширения функциональности этой игры (например, играют два человека, играют две программы и т.п.).

**Вариант 3.** Программа "Календарь-ежедневник": на заданную дату в заданное время реализовать возможность запланировать событие. Реализовать функции создания, редактирования, удаления событий. Предложите и реализуйте  расширения функциональности этой программы (можно ориентироваться на функциональность ежедневника OutLook и аналогичных приложений).

**Вариант 4.** Игра "Жизнь": имеется поле, поделенное на клетки. Каждая клетка на этой поверхности может находиться в двух состояниях – "живая" или "мёртвая". Распределение "живых" клеток в начале игры называется первым поколением. Каждое следующее поколение (шаг) рассчитывается на основе предыдущего по таким правилам: пустая ("мёртвая") клетка с $Nx$ "живыми" клетками-соседями оживает; если у "живой" клетки есть не менее $(Nx-1)$ "живых" соседей, то эта клетка продолжает жить, если "живых" соседей меньше $(Nx-1)$ или больше $(Nx+1)$, то клетка "умирает" (от "одиночества" или от "перенаселённости"). Предложите и реализуйте  расширения функциональности этой игры (например, игрок сам устанавливает правила взаимодействия клеток, расставляет "живые" клетки первого поколения, конфигурации генерируются случайно и т.п.).

**Вариант 5.** Устный счет: используя арифметические операции умножения, сложения, вычитания, деления, возведения в степень, программа случайно генерирует арифметическое выражение, используя в случае необходимости скобки, вычисляет его значение и предлагает человеку на время также выполнить вычисление.  Предложите и реализуйте расширения функциональности этой игры (например, регулируется длина выражения, используемые арифметические операции, предусматривается система бонусов, если игрок угадывает решение и т.п.).

**Вариант 6.** Игра "Cапёр": на поле 10x10 программа случайно размещает фиксированное количество мин. Требуется разминировать поле. Игроку разрешается открывать любую клетку или помечать ее признаком - заминировано. После каждой открытой клетки программа сообщает для каждой смежной с ней клетки,  сколько смежных с ней мин. Если игрок открыл клетку с миной, он проиграл. Если он обезвредил все мины, он выиграл. Предложите и реализуйте  расширения функциональности этой игры (например, игрок сам выбирает размерность поля, количество мин, количество ошибок и т.п.).

**Вариант 7.** Игра "Пятнашки": в клетках поля 4x4 размещаются фишки от 1 до 15, одна клетка остается пустой. В пустую клетку можно перемещать любую из соседних с ней клеток. Необходимо реализовать перемещение фишек, визуализацию поля. Требуется вернуть искомую нумерацию. Предложите и реализуйте расширения функциональности этой игры (например, программа сама тасует фишки, сама пытается собрать требуемую конфигурацию и т.п.).

**Вариант 8.** Игра "Лабиринт": играющий перемещается в двухмерном пространстве по помещениям здания, план которого играющему неизвестен. Начиная с произвольного помещения, путешественник должен найти выход из здания. Каждое помещение может иметь четыре двери: север, восток, юг, запад и соединяться с другими помещениями. План здания необходимо сгенерировать случайно. Порядок следования помещений в списке должен быть произвольным. Находясь в N–ом помещении, игрок может получить подсказку о правильном направлении движения, если верно выполнит тестовое задание по теме "Системное программирование". Предложите и реализуйте расширения функциональности этой игры (например, автоматическое или ручное прохождение роботом лабиринта, несколько выходов и т.п.).

**Вариант 9.** Программа "Моделирование компьютера": имеется специальный регистр - аккумулятор, в который помещается результат арифметических операций, различных операций сравнения. Имеется оперативная память из 100 ячеек, куда можно записать любое целое число. Программа состоит из команд, каждая команда - это четырехзначное число. Первые две цифры - это код команды, которую нужно выполнить: 10 - вводится слово с терминала в указанное место памяти; 11 - выводится слово на терминал из указанного адреса памяти; 20 - в аккумулятор помещается слово из указанного адреса памяти; 21 - в память помещается слово из аккумулятора; 30 - сложение слова аккумулятора и слова из указанного места в памяти (результат остается в аккумуляторе); 31  - вычитание слова аккумулятора и слова из указанного места в памяти (результат остается в аккумуляторе); 32 - деление слова аккумулятора на слово из указанного места памяти; 40 - переход по указанному адресу памяти; 41  - переход по указанному адресу памяти, если в аккумуляторе находится отрицательное число; 42 - переход по указанному адресу памяти, если в аккумуляторе находится нуль; 43 - останов, выполняется при завершении программой своей работы. Необходимо реализовать ввод и выполнение программы. Предложите и реализуйте расширения функциональности этой программы (например, можно расширить набор команд, пытаться программой писать программы самой себе, составить несколько полезных программ т.п.). 

**Вариант 10.** Программа "крестики-нолики": в любую пустую клетку на поле 3x3 программа может ставить нолик, человек - крестик. Выигрывает тот, кто первым составит горизонтальный, вертикальный или диагональный ряд своих символов (крестиков или ноликов). Предложите и реализуйте расширения функциональности этой игры (например, меняется размер поля, программа следует какой-то стратегии в игре, программы играют друг с другом и т.п.). 

**Вариант 11.** Программа "Шарики": в игровом пространстве летают красные, синии шарики (некоторые объекты), отражаясь от краев пространства по принципу - угол падения $\alpha$ равен углу отражения $\pi-\alpha$. У игрока имеется объект, который может перемещаться в игровом пространстве вправо, влево, вверх, вниз путем нажатия соответствующих клавиш. Первоначально у объекта игрока есть 100 единиц жизни, при столкновении с красным шариком жизненный аккумулятор уменьшается на единицу, при столкновении с синим - увеличивается на единицу. Предусмотреть расширение функциональности игры: разные типы шариков - разные бонусы, несколько игроков, остановить - запустить - сохранить игру и т.д.

**Вариант 12.** Программа "Теннис": в игровом пространстве летает шарики (объекты), отражаясь от краев пространства по принципу -  угол падения $\alpha$ равен углу отражения $\pi-\alpha$. У игрока имеется бита, которая может перемещаться вправо, влево путем нажатия соответствующих клавиш и находится внизу игрового пространства. При соударении с битой объект отражается от нее по принципу  угол падения равен углу отражения. Первоначально у игрока есть 100 единиц жизни, при пропуске шарика жизненный аккумулятор уменьшается на единицу, при отбое шарика - увеличивается на единицу. Предусмотреть расширение функциональности игры: несколько шариков с разными бонусами, несколько игроков, остановить - запустить - сохранить игру и т.д.

**Вариант 13.** Програма "Распределенное хранилище": с использованием сокетов реализовать кластерную архитектуру: главный сервер (отдельный процесс на выделенном порте) и его реплики (другие процессы, каждый со своим портом). Кластер хранит информацию по принципу ключ-значение. Главный сервер отвечает за запись информации в кластер, реплики реплицируют (случайно дублируют информацию) с главного компьютера и отвечают на чтение. Выбор, какая реплика должна ответить на чтение осуществляет главный сервер. Если главный сервер падает, реплики выбирают себе новый главный сервер, если его обратно поднимают, он становится простой репликой. Предусмотреть расширение функциональности: умные алгоритмы репликации, смоделировать генератор параллельных циклов чтения, записи и т.д.

**Вариант 14.** Программа "Банк": разработать сетевое клиент-серверное приложение для организации банковских счетов и доступа к ним пользователей. Программа должна позволять устанавливать соединение с несколькими клиентами (до 10), и реализовывать функционал перевода денежных средств, снятия, пополнения, взятие кредита, погашение кредита, режим авторизации. Предусмотреть расширение функциональности: коммисия в пользу банка при переводах, при погашении кредитов, хранение всей информации в файлах, работу нескольких банков друг с другом.

**Вариант 15.** Программа "Стеганография": реализовать алгоритм скрытой передачи текстовой информации в изображении, в программе должны быть функции шифровальщика и дешифровальщика. При этом изображение не должно заметно искажаться.  Предусмотреть расширение функциональности программы: вместо изображения использовать звуковой файл или наоборот скрывать звуковую или графическую информацию и т.д.

**Вариант 16.** Программа "SIMD": с использованием технологии single instruction, multiple data реализовать ускоренные алгоритмы вычисления среднего значения большого массива чисел. Предусмотреть расширение функциональности программы: сортировка элементов массива чисел заданной разрядности, нахождения заданного квантиля без полной сортировки и т.д.  

**Вариант 17.** Программа "Платформа тестирования": пользователю случайным образом предлагается ряд вопросов и варианты ответа по
дисциплине «Системное программирование». Предлагается ввести и пройти тест. По окончанию тестирования создается статистика прохождения, вывод результатов, количество правильно и неправильно выбранных ответов, оценка. Необходимо предусмотреть возможность составить тест,
вводя вопрос и варианты ответа. Предусмотреть расширение функциональности программы: реализовать удаленное тестирование с использованием сокетов, расширить тип тестов (выбор нескольких правильных ответов) и т.д.


**Вариант 18.** Программа "Файловый менеджер": программа в диалоговом режиме реализует все основные команды работы с файлами: создание, переименование, копирование, перемещение, создание жесткой, символической ссылки, изменение текущего каталога, просмотр свойств файла: размер, дата изменения и т.д. Предусмотреть расширение функциональности программы: с использование расширенных атрибутов файла и выбранного алгоритма хэширования добавить атрибут, хранящий хэш файла, реализовать доступ к файловой системе удаленного компьютера (используя сокеты) и т.д.

**Вариант 19.** Программа "Текстовый редактор": реализовать консольный текстовый редактор (по аналогии с nano), который должен предусматривать перемещение курсора, печать символов, копирование, вставка, вырезание, сохранение текста, загрузка текста. **Вариант 20.** Программа "Оценка производительности алгоритмов сортировки": реализовать на языке ассемблера и на языке С несколько алгоритмов сортировки: быстрая сортировка Хоара, сортировка слиянием, пирамидальная сортировка, поразрядная сортировка. Оценить быстродейсвие алгоритмов на разных числовых последовательностях. Предусмотреть расширение функциональности программы: расширить перечень алгоритмов сортировки - битонная сортировка, сортировка вставками, реализовать параллельные варианты алгоритмов и т.д.

**Вариант 21.** Программа "Искусственная жизнь. Эволюция" — на поле существуют существа (агенты) с простым генетическим кодом. У них есть энергия. Они могут двигаться, есть еду (растет случайно), размножаться и умирать. Реализовать пошаговую симуляцию. Расширения: мутации при размножении, наследование признаков (скорость, дальность зрения), введение хищников, визуализация графа родственных связей, сохранение и загрузка состояний популяции.

**Вариант 22.** Программа "Генератор паролей" — программа создает надежные пароли заданной длины, используя буквы разного регистра, цифры и специальные символы. Пользователь может указать желаемую длину и набор символов. Расширения: оценка стойкости сгенерированного пароля, сохранение паролей в зашифрованном хранилище с мастер-паролем, генерация произносимых паролей, проверка, не попал ли пароль в базы утекших данных (через API), автоматическая смена паролей по расписанию.

**Вариант 23.** Программа "Телефонная книга" — консольное приложение для хранения контактов: имя, фамилия, несколько номеров телефонов, email, день рождения. Реализовать добавление, удаление, поиск, редактирование. Расширения: импорт/экспорт vCard, напоминания о днях рождения, группировка контактов (друзья, коллеги), поиск дубликатов и т.д.

**Вариант 24.** Программа "Генератор QR-кодов" — преобразует введенный текст или URL в QR-код и выводит его на экран (псевдографикой в консоли или сохраняет как изображение). Расширения: декодирование QR-кода из изображения, генерация разных версий  и т.д.

**Вариант 25.** Программа "Система голосования" — реализовать клиент-серверное приложение для проведения голосований. Сервер хранит список вопросов и вариантов ответов. Клиенты подключаются, авторизуются и голосуют. Расширения: анонимное голосование с использованием слепых подписей, подсчет результатов в реальном времени с выводом графиков, создание новых голосований администратором, тайное голосование с последующей расшифровкой.

**Вариант 26.** Программа "Поисковик дубликатов файлов" — сканирует указанную директорию и ищет дубликаты файлов (по имени, размеру, хэш-сумме). Предлагает пользователю удалить или переместить дубликаты. Расширения: fuzzy-поиск похожих изображений (не точных копий), выборочное исключение папок из сканирования, многопоточное сканирование.

**Вариант 27.** Программа "Онлайн-библиотека" — клиент-серверное приложение для учета книг. Сервер хранит каталог книг, информацию о читателях и выданных экземплярах. Клиенты могут искать книги, бронировать и сдавать их. Расширения: электронные версии книг (загрузка/скачивание), рейтинги и отзывы читателей, автоматические напоминания о просрочке, штрафы за задержку, очередь на популярные книги.

**Вариант 28.** Программа "Симулятор работы процессора" — программная модель простого учебного процессора с ограниченным набором команд (ADD, SUB, LOAD, STORE, JMP). Реализовать ассемблер для этого процессора и среду исполнения с пошаговым режимом и просмотром регистров. Расширения: добавление стека и поддержки подпрограмм (CALL, RET), моделирование прерываний, конвейеризация, кэш-память, визуализация работы АЛУ.

**Вариант 29.** Программа "Сетевой чат с шифрованием" — многопользовательский чат с центральным сервером. Все сообщения передаются в зашифрованном виде (например, с использованием асимметричного шифрования для обмена ключами и симметричного для трафика). Расширения: поддержка приватных сообщений, отправка файлов, создание комнат (каналов), история сообщений с поиском и т.д.

**Вариант 30.** Программа "Морзянка" — преобразование текста в азбуку Морзе и обратно. Поддержка световой сигнализации (мигание экрана разных светов с разной длительностью), а также текстовой генерации. Расширения: генератор случайных последовательностей для тренировки, семафорная азбука, самостоятельное залание правил и т.д.
